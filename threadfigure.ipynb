```python
import threadFigure
# For the Cube Figure
def cube_csa(a):
    # Cube has no curved surface; returning lateral surface area
    return 4 * a * a

def cube_tsa(a):
    return 6 * a * a

def cube_volume(a):
    return a ** 3

# For the Cuboid figure
def cuboid_csa(l, b, h):
    return 2 * h * (l + b)

def cuboid_tsa(l, b, h):
    return 2 * (l*b + b*h + h*l)

def cuboid_volume(l, b, h):
    return l * b * h

 # For the Cylinder
def cylinder_csa(r, h):
    return 2 * math.pi * r * h

def cylinder_tsa(r, h):
    return 2 * math.pi * r * (h + r)

def cylinder_volume(r, h):
    return math.pi * r**2 * h


#For the Cone Figure
def cone_csa(r, l):
    return math.pi * r * l

def cone_tsa(r, l):
    return math.pi * r * (l + r)

def cone_volume(r, h):
    return (1/3) * math.pi * r**2 * h


# For the Sphere Figure
def sphere_csa(r):
    return 4 * math.pi * r**2   # same as TSA

def sphere_tsa(r):
    return 4 * math.pi * r**2

def sphere_volume(r):
    return (4/3) * math.pi * r**3

# For the Hemisphere Figure
def hemisphere_csa(r):
    return 2 * math.pi * r**2

def hemisphere_tsa(r):
    return 3 * math.pi * r**2

def hemisphere_volume(r):
    return (2/3) * math.pi * r**3

import threadFigure as g

print("Cube Volume:", g.cube_volume(3))
print("Cylinder CSA:", g.cylinder_csa(7, 10))
print("Sphere TSA:", g.sphere_tsa(5))



```
